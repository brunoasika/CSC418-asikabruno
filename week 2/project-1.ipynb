```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

def enhancement():
    user_images = { 'bruno': (1001, r'C:\Users\hp\Documents\CSC418-asikabruno\week 2\img\bruno.jpg'), 
     'david': (2002, r'C:\Users\hp\Documents\CSC418-asikabruno\week 2\img\david.jpg'), 
     'eze': (3003, r'C:\Users\hp\Documents\CSC418-asikabruno\week 2\img\eze.jpg'), 
     'francis': (4004, r'C:\Users\hp\Documents\CSC418-asikabruno\week 2\img\francis.jpg'), 
     'kaobi': (5005, r'C:\Users\hp\Documents\CSC418-asikabruno\week 2\img\kaobi.jpg'), 
     'lala': (6006, r'C:\Users\hp\Documents\CSC418-asikabruno\week 2\img\lala.jpg'),
     'lotanna': (7007, r'C:\Users\hp\Documents\CSC418-asikabruno\week 2\img\lotanna.jpg'), 
     'oyinda': (8008, r'C:\Users\hp\Documents\CSC418-asikabruno\week 2\img\oyinda.jpg'), 
     'pese': (9009, r'C:\Users\hp\Documents\CSC418-asikabruno\week 2\img\pese.jpg'), 
     'zion': (1010, r'C:\Users\hp\Documents\CSC418-asikabruno\week 2\img\zion.jpg') } 
    
    username = input("Enter username (first name): ") 
    password = int(input("Enter password (matric number): ")) 
    
    if username in user_images: 
        matric, image_path = user_images[username]
        expected_password = matric 
        if password == expected_password: 
            try:
                Option = input("Select an enhancement you would like to see. (Contrast, Addition, Subtraction, Sharpeness, MedianBlur, Noise, Scaling, Transformation):") 
                if Option == "Contrast": 
                    image = cv2.imread(image_path)
                    plt.subplot(1,2,1)
                    plt.title(f'original {username}')
                    plt.imshow(image)
                    brightness = 5
                    contrast = 1.5
                    image2 = cv2.addWeighted(image, contrast, np.zeros(image.shape,image.dtype), 0, brightness)
                    cv2.imwrite(f'img/{username}_contrast.jpg', image2)
                    plt.subplot(1,2,2)
                    plt.title(f'contrasted {username}')
                    plt.imshow(image2)
                    plt.show() 
                    
                elif Option == "Addition":
                    image1 = cv2.imread('img/sst.jpg')
                    image = cv2.imread(image_path)
                    image1 = cv2.resize(image1, (500,400))
                    image = cv2.resize(image, (500,400))
                    addImage = cv2.addWeighted(image1, 0.5, image, 0.6, 0)
                    cv2.imshow('weighted Image', addImage)
                    if cv2.waitKey(0) & 0xff == 27:
                        cv2.destroyAllWindows()
                        
                elif Option == "Subtraction":
                    image1 = cv2.imread('img/sst.jpg')
                    image = cv2.imread(image_path)
                    image1 = cv2.resize(image1, (500,400))
                    image = cv2.resize(image, (500,400))
                    sub = cv2.subtract(image1, image)
                    plt.subplot(1,2,2)
                    plt.imshow(sub)
                    plt.show()
                    if cv2.waitKey(0) & 0xff == 27:
                        cv2.destroyAllWindows()
                
                elif Option == "MedianBlur":
                    image = cv2.imread(image_path)
                    plt.subplot(1, 2, 1)
                    plt.title("original")
                    plt.imshow(image)
                    filtered_image = cv2.medianBlur(image, 15)
                    cv2.imwrite('img/median-Blur.jpg', filtered_image)
                    plt.subplot(1, 2, 2)
                    plt.title("Median Blur")
                    plt.imshow(filtered_image)
                    plt.show()
                    
                
                elif Option == "Sharpeness":
                    image = cv2.imread(image_path)
                    plt.subplot(1,2,1)
                    plt.title(f'original {username}')
                    plt.imshow(image)
                    kernel = np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])
                    sharpened_image = cv2.filter2D(image, -1, kernel)
                    cv2.imwrite(f'img/{username}_Sharpened.jpg', sharpened_image)
                    plt.subplot(1,2,2)
                    plt.title(f'sharpened {username}')
                    plt.imshow(sharpened_image)
                    plt.show()
                        
                elif Option == "Noise":
                    image = cv2.imread(image_path)
                    plt.subplot(1,2,1)
                    plt.title(f'original {username}')
                    plt.imshow(image)
                    filtered_image = cv2.medianBlur(image, 15)
                    cv2.imwrite(f'img/{username}_Noiseless.jpg', filtered_image)
                    plt.subplot(1,2,2)
                    plt.title(f'noiseless {username}')
                    plt.imshow(filtered_image)
                    plt.show()

                elif Option == "Scaling":
                    image = cv2.imread(image_path)
                    plt.subplot(1,2,1)
                    plt.title(f'original {username}')
                    plt.imshow(image)
                    scaled_image = cv2.resize(image, None, fx=2, fy=2)
                    cv2.imwrite(f'img/{username}_Scaled.jpg', scaled_image)
                    plt.subplot(1,2,2)
                    plt.title(f'scaled {username}')
                    plt.imshow(scaled_image)
                    plt.show()

                elif Option == "Transformation":
                    image = cv2.imread(image_path)
                    plt.subplot(1,2,1)
                    plt.title(f'original {username}')
                    plt.imshow(image)
                    inverse_image = 255 - image
                    cv2.imwrite(f'img/{username}_Transformed.jpg', inverse_image)
                    plt.subplot(1,2,2)
                    plt.title(f'inverse color {username}')
                    plt.imshow(inverse_image)
                    plt.show()

                else:
                    print("Invalid Selection")
            except Exception as e: 
                print('Error:', e)

        else:
            print("Password Not seen")

    else:
        print("Authentication failed. User not found.")
        
enhancement()
```
