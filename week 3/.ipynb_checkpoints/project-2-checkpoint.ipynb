```python
tk_image = None

def main():
    global tk_image
    root = tk.Tk()
    root.title("Artwork Viewer")

    images = {
        'Traditional': {
            'Yoruba humn face': r'C:\Users\hp\Documents\CSC418-asikabruno\week 3\trad\t1',
            'Benin Head': r'C:\Users\hp\Documents\CSC418-asikabruno\week 3\trad\t2'
        },
        'Modern': {
            'Untitled': r'C:\Users\hp\Documents\CSC418-asikabruno\week 3\modern\m1',
            'Untitled no image': r'C:\Users\hp\Documents\CSC418-asikabruno\week 3\modern\m2'
        },
        'Contemporary': {
            'Flexibility': r'C:\Users\hp\Documents\CSC418-asikabruno\week 3\contemp\c1',
            'Durbar': r'C:\Users\hp\Documents\CSC418-asikabruno\week 3\contemp\c2'
        }
    }

    username = input("Enter your username: ")
    password = input("Enter your password: ")
    category = input("Enter the category you want to see (Traditional, Modern, Contemporary): ")

    if isinstance(username, str) and isinstance(password, str):
        if category in images:
            print(f"Artworks in the {category} category:")
            for artwork_name, artwork_path in images[category].items():
                print(artwork_name, artwork_path)
            artwork = input("Enter the artwork_name of the artwork you want to see: ")
            if artwork in images[category]:
                enhancement = input("Select enhancement you would like to see on the artwork (Contrast, Addition, Subtraction, Sharpeness, MedianBlur, Noise, Scaling, Transformation): ")
                image_path = images[category][artwork]
                display_image(root, image_path, enhancement)
                root.mainloop()  # Start the Tkinter event loop
            else:
                print("Artwork not found in the selected category.")
        else:
            print(f"No artworks found in the {category} category.")
    else:
        print("Invalid username or password.")

def display_image(root, image_path, enhancement):
    global tk_image
    
    apply_enhancement(image_path, enhancement)
    
    image = Image.open(image_path)
    image.thumbnail((400, 400))
    tk_image = ImageTk.PhotoImage(image)
    
    for widget in root.winfo_children():
        widget.destroy()
    
    label = Label(root, image=tk_image)
    label.image = tk_image
    label.pack(padx=10, pady=10)


def apply_enhancement(image_path, enhancement):
    if enhancement == "Contrast":
        contrast(image_path)
    elif enhancement == "Addition":
        addition(image_path)
    elif enhancement == "Subtraction":
        subtraction(image_path)
    elif enhancement == "Sharpeness":
        sharp(image_path)
    elif enhancement == "MedianBlur":
        blur(image_path)
    elif enhancement == "Noise":
        noise(image_path)
    elif enhancement == "Scaling":
        scale(image_path)
    elif enhancement == "Transformation":
        transform(image_path)
    else:
        print("Invalid selection")

def contrast(image_path):
    image = cv2.imread(image_path)
    plt.subplot(1,2,1)
    plt.title(f'original {username}')
    plt.imshow(image)
    brightness = 5
    contrast = 1.5
    image2 = cv2.addWeighted(image, contrast, np.zeros(image.shape,image.dtype), 0, brightness)
    cv2.imwrite(f'img/{username}_contrast.jpg', image2)
    plt.subplot(1,2,2)
    plt.title(f'contrasted {username}')
    plt.imshow(image2)
    plt.show() 

def addition(image_path):
    image1 = cv2.imread('img/sst.jpg')
    image = cv2.imread(image_path)
    image1 = cv2.resize(image1, (500,400))
    image = cv2.resize(image, (500,400))
    addImage = cv2.addWeighted(image1, 0.5, image, 0.6, 0)
    cv2.imshow('weighted Image', addImage)
    if cv2.waitKey(0) & 0xff == 27:
        cv2.destroyAllWindows()

def subtraction(image_path):
    image1 = cv2.imread('img/sst.jpg')
    image = cv2.imread(image_path)
    image1 = cv2.resize(image1, (500,400))
    image = cv2.resize(image, (500,400))
    sub = cv2.subtract(image1, image)
    plt.subplot(1,2,2)
    plt.imshow(sub)
    plt.show()
    if cv2.waitKey(0) & 0xff == 27:
        cv2.destroyAllWindows()

def blur(image_path):
    image = cv2.imread(image_path)
    plt.subplot(1, 2, 1)
    plt.title("original")
    plt.imshow(image)
    filtered_image = cv2.medianBlur(image, 15)
    cv2.imwrite('img/median-Blur.jpg', filtered_image)
    plt.subplot(1, 2, 2)
    plt.title("Median Blur")
    plt.imshow(filtered_image)
    plt.show()

def sharp(image_path):
    image = cv2.imread(image_path)
    plt.subplot(1, 2, 1)
    plt.title("original")
    plt.imshow(image)
    filtered_image = cv2.medianBlur(image, 15)
    cv2.imwrite('img/median-Blur.jpg', filtered_image)
    plt.subplot(1, 2, 2)
    plt.title("Median Blur")
    plt.imshow(filtered_image)
    plt.show()

def noise(image_path):
    image = cv2.imread(image_path)
    plt.subplot(1,2,1)
    plt.title(f'original {username}')
    plt.imshow(image)
    filtered_image = cv2.medianBlur(image, 15)
    cv2.imwrite(f'img/{username}_Noiseless.jpg', filtered_image)
    plt.subplot(1,2,2)
    plt.title(f'noiseless {username}')
    plt.imshow(filtered_image)
    plt.show()

def scale(image_path):
    image = cv2.imread(image_path)
    plt.subplot(1,2,1)
    plt.title(f'original {username}')
    plt.imshow(image)
    scaled_image = cv2.resize(image, None, fx=2, fy=2)
    cv2.imwrite(f'img/{username}_Scaled.jpg', scaled_image)
    plt.subplot(1,2,2)
    plt.title(f'scaled {username}')
    plt.imshow(scaled_image)
    plt.show()

def transform(image_path):
    image = cv2.imread(image_path)
    plt.subplot(1,2,1)
    plt.title(f'original {username}')
    plt.imshow(image)
    inverse_image = 255 - image
    cv2.imwrite(f'img/{username}_Transformed.jpg', inverse_image)
    plt.subplot(1,2,2)
    plt.title(f'inverse color {username}')
    plt.imshow(inverse_image)
    plt.show()
    
    
main()
```
