```python
import numpy as np
import matplotlib.pyplot as pt
import cv2 as cv

```

```python
img = cv.imread('img/sst.jpg', 0)
rows, cols = img.shape

```

```python
#plot the original image
pt.subplot(1, 2, 1)
pt.title("Original")
pt.imshow(img)

```

```python
image_rotation = cv.warpAffine(img, cv.getRotationMatrix2D((cols/2, rows/2), 30, 0.6), (cols, rows))


#plot the translated image
pt.subplot(1, 2, 2)
pt.title("rotated image")
pt.imshow(image_rotation)

cv.imshow('image', image_rotation)
cv.imwrite('img/rotation_out.jpg', image_rotation)
cv.waitKey(0)
cv.destroyAllWindows()
```
