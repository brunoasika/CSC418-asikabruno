```python
import numpy as np
import matplotlib.pyplot as pt
import cv2 as cv

img = cv.imread('img/sst.jpg', 0)
rows, cols = img.shape

#plot the original image
pt.subplot(1, 2, 1)
pt.title("Original")
pt.imshow(img)

cropped_img = img[450:700, 200:500]


#plot the cropped image
pt.subplot(1, 2, 2)
pt.title("cropped image")
pt.imshow(cropped_img)

cv.imwrite('img/cropped_out.jpg', cropped_img)
cv.waitKey(0)
cv.destroyAllWindows()
```
